## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd, numpy as np
df = pd.read_csv("/content/content_refresh_anonymized.csv")

# Number 1: does search volume predict actual traffic?
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation search_volume vs impressions_90d: {corr:.3f}")

# Number 2: CTR by position tier
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print("\nMean CTR by position tier:")
print(ctr_by_pos.round(4))

# Number 3: word count vs trend direction
wc = df.groupby("trend_direction")["word_count"].median()
print("\nMedian word count by trend direction:")
print(wc.round(0))
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

```python
print("""
What I can claim (observed, directional, decision-support):

1. Search volume vs impressions: the correlation is 0.001 — essentially
   zero. This means, in this sample, a keyword's search volume does NOT
   reliably predict how many impressions a page actually gets. This is a
   directional finding: a content team should not assume "high search
   volume keyword = guaranteed traffic."

2. Position tier vs CTR: there is a clear, consistent pattern — CTR drops
   from 0.355 at page_1 down to 0.055 for deep-ranked pages. This is a
   strong, well-established relationship, not just a coincidence in this
   sample. It supports using position tier as a key signal when deciding
   which pages need attention.

3. Word count vs trend direction: median word count is roughly similar
   across down (2909), flat (2698), stable (2912), and up (2848) pages,
   with only 'new' pages noticeably lower (2239, likely because they simply
   haven't been expanded yet). This suggests word count alone does not
   clearly separate declining pages from growing ones.

What I cannot claim:

- I cannot claim these numbers prove causation. A weak correlation (search
  volume) doesn't prove there's truly no relationship anywhere -- only that
  it isn't reliable in this sample. A near-equal median (word count) doesn't
  prove length never matters -- only that it isn't the deciding factor here.
- I cannot claim this reflects how Google's actual ranking algorithm works.
- I cannot claim any of this is causal (e.g., that ranking position CAUSES
  higher CTR in a way I could intervene on) -- these are observed
  associations from historical data, not a controlled experiment.
- This is decision-support: it should inform where a reviewer looks first,
  not a guarantee about what will happen if something is changed.
""")
```
