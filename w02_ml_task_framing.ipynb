## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
print("""
Primary metric: feature importance ranking from a Random Forest classifier
-- which signals get weighted highest when predicting decline.

Supporting metric: ROC AUC on held-out data, to confirm the model found a
real, generalizable pattern rather than noise. If ROC AUC sits near 0.5,
the "important" features aren't actually meaningful -- they're random.

I'm not using Precision@K here since this lane isn't producing a review
queue (that's Lane 2's job) -- it's about which signals deserve attention
at all.
""")
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd, numpy as np
df = pd.read_csv("/content/content_refresh_anonymized.csv")

print("Unit of analysis: one row = one content page")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
df.head(3)
```

```python

df["is_declining_label"] = df["trend_direction"].str.lower().eq("down").astype(int)
print(df["is_declining_label"].value_counts())
print(f"\nDeclining rate: {df['is_declining_label'].mean():.3f}")
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
print("""
A fixed rule can only weigh 1-2 signals by hand -- like Week 2's
"stale x visible" rule, a human guess at which two signals mattered most.
But this data has 10+ candidate signals (position, CTR, word count,
freshness, age, content type, etc.), and their relative importance isn't
obvious from eyeballing one-at-a-time bar charts, like Week 1's approach.

A model like Random Forest weighs all these signals simultaneously, ranks
them by actual contribution, and can catch combinations a human wouldn't
manually test. That's a systematic signal audit instead of an ad-hoc
guess -- which is what Lane 1 is actually asking for.
""")
```
